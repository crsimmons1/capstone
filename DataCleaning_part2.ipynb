### Cardinality and NA Values per Column

```python
def cardinality_categorical(df):
    n_records = len(df)
    for column in df.select_dtypes([object]):
        print("{} | uniques/records: {:.3f} | Minimum observations: {:.3f}".format(
            column, 
            len(df[column].unique())/n_records,
            df[column].value_counts().min()
        ))        

cardinality_categorical(baltimore)
```

```python
n_records = len(baltimore)
def missing_values_df(df):
    for column in df:
        print("{} | {} | {}".format(
            column, len(df[df[column].isnull()]) / (1.0*n_records), df[column].dtype
        ))

missing_values_df(baltimore)
```

```python
from scipy import stats
import numpy as np

def outliers_col(df):
    for column in df:
        if df[column].dtype != np.object:
            n_outliers = len(df[(np.abs(stats.zscore(df[column])) > 3)& \
                  (df[column].notnull())
                 ])
            print("{} | {} | {}".format(
                df[column].name,
                n_outliers,
                df[column].dtype
        ))

outliers_col(baltimore)
```

### Variable for Time - Daypart

```python
baltimore["Hour"]= baltimore["CrimeTime"].astype(str).str[:2].str.replace(r':', '')
```

```python
# delete all rows for which column 'na' is in for 'HOUR'
baltimore['Hour'].astype(str)

indexNames = baltimore[ baltimore['Hour'] == 'na' ].index
baltimore = baltimore.drop(indexNames,inplace=False)
baltimore['Hour'].value_counts()
# convert hour into type 'int' to do comparison
baltimore['Hour'] = baltimore['Hour'].astype(int)
```

```python
# Assigned times into different parts of the day and created a variable called 'DAYPART'
def daypart(time):
    if ((time >= 0) and (time <=6)):
        return 'Dawn'
    elif ((time > 6) and (time <= 9)):
        return 'Early morning'
    elif ((time > 9) and (time <=12)):
        return 'Morning/Noon'
    elif ((time > 12) and (time <=17)):
        return 'Afternoon'
    elif ((time > 17) and (time <=21)):
        return 'Evening'
    elif ((time > 21) and (time <=23)):
        return 'Night'

baltimore['DayPart'] = baltimore['Hour'].apply(daypart)
```

```python
modes = date.groupby("CrimeDate")["Description"].agg(pd.Series.mode).to_frame()
```

```python
modes["Year"] = date.groupby("CrimeDate")["year"].agg(pd.Series.mode).to_frame()
modes["Month"] = date.groupby("CrimeDate")["month"].agg(pd.Series.mode).to_frame()
modes["Day"] = date.groupby("CrimeDate")["day"].agg(pd.Series.mode).to_frame()
```

```python
modes["ViolentCount"] = date.groupby("CrimeDate")["Violent"].agg(['sum'])
modes["NonViolentCount"] = date.groupby("CrimeDate")["Theft"].agg(['sum'])
```

```python
modes["Temperature"] = date.groupby("CrimeDate")["TAVG"].agg(pd.Series.mode).to_frame()
modes["Percipitation"] = date.groupby("CrimeDate")["PRCP"].agg(pd.Series.mode).to_frame()
modes["Snow"] = date.groupby("CrimeDate")["SNOW"].agg(pd.Series.mode).to_frame()
modes["SnowDepth"] = date.groupby("CrimeDate")["SNWD"].agg(pd.Series.mode).to_frame()
modes["Unemployment"] = date.groupby("CrimeDate")["Baltimore.City.Unemployment.Percent"].agg(pd.Series.mode).to_frame()
```

```python
#modes["Weapon"] = baltimore.groupby("CrimeDate")["Weapon"].agg(pd.Series.mode).to_frame()
modes["Larceny"] = date.groupby("CrimeDate")["LARCENY"].agg(['sum'])
modes["AutoTheft"] = date.groupby("CrimeDate")["AUTO THEFT"].agg(['sum'])
modes["Assault"] = date.groupby("CrimeDate")["ASSAULT"].agg(['sum'])
modes["Burglary"] = date.groupby("CrimeDate")["BURGLARY"].agg(['sum'])
modes["Robbery"] = date.groupby("CrimeDate")["ROBBERY"].agg(['sum'])
modes["Arson"] = date.groupby("CrimeDate")["ARSON"].agg(['sum'])
modes["Homicide"] = date.groupby("CrimeDate")["HOMICIDE"].agg(['sum'])
modes["Rape"] = date.groupby("CrimeDate")["RAPE"].agg(['sum'])
modes["Shooting"] = date.groupby("CrimeDate")["SHOOTING"].agg(['sum'])
```

```python
modes["Dawn"] = date.groupby("CrimeDate")["Dawn"].agg(['sum'])
modes["Morning/Noon"] = date.groupby("CrimeDate")["Morning/Noon"].agg(['sum'])
modes["Night"] = date.groupby("CrimeDate")["Night"].agg(['sum'])
modes["Evening"] = date.groupby("CrimeDate")["Evening"].agg(['sum'])
modes["Early morning"] = date.groupby("CrimeDate")["Early morning"].agg(['sum'])
modes["Afternoon"] = date.groupby("CrimeDate")["Afternoon"].agg(['sum'])
```
